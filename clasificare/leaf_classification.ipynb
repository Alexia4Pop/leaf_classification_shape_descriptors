```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the data
df = pd.read_csv('leaf.csv')

# 2. Setup the plot
plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'orange', 'purple']
species = df['Species'].unique()

# 3. Plot each species
for i, s in enumerate(species):
    subset = df[df['Species'] == s]
    plt.scatter(subset['Elongation'], subset['Circularity'],
                label=s, c=colors[i], edgecolors='black', s=80)

# 4. Labeling
plt.xlabel('Elongation (Shape Factor)')
plt.ylabel('Circularity (Roundness)')
plt.title('Leaf Classification: Feature Space Plot')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
